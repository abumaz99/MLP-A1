## Exploratory Analysis
### Normality
#### Checking Normality
- Use Shapiro Wilk's Test to verify normality of the variables.
- Regression models assume normal distribution, therefore if any variables are non-normal we should exclude them or perform some transformations to make them normal 



#### Addressing Non-Normality

##### Outlier Analysis
- Detecting and removing outliers makes data more normal, should however only done with care



##### Log Transformation
- Log transforms of the data bring in the tails and make data more normal 



### Checking Correlation
- Run a standard corrleation matrix across the feature space
- Correlated data is fine as long as correlation is not above 0.9
- If two features are highly correlated we exclude one or address this in interpretation. Possibly LASSO deals with this automatically as well but good to be aware beforehand



### Standard Scaling
- Definitely good practice to scale continuous variables to mean zero and standard deviation = 1 to bring them on the same scale

### Polynomial Feature Selection
Can we improve on the above CV RMSE by reducing the complexity of our model? 

Typically, there are two ways of going about this:
- LASSO Regularization to shrink coefficients to zero
- Stepwise Selection
    - Backwards Stepwise Selection starts with the full model and sequentially reduces it down until it no longer improves
        - Typically produces a larger model
    - Forwards Stepwise Selection starts with the null model (only intercept) and sequentially adds features
        - Typically produces smaller model than backwards stepwise
